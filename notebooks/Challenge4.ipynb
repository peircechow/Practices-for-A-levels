```python
'''
The Fibonacci sequence was originally used as a basic model for rabbit population growth.
The Fibonacci sequence goes like this:

0,1,1,2,3,5,8,13,...

The nth term is the sum of the previous two terms.
In the above example, the next term would be 21 (8+13).

Create a Fibonacci sequence generator up to the 50th term. 
The program will then prompt the user for n (up to 50).
E.g
Enter n: 7
Fibonacci number is 8
'''
def fib(n):
    if n == 0:
        return 0
    elif n== 1:
        return 1
    else:
        return fib(n-1) + fib(n-2)
    
    
n = int(input("Enter n:"))
print("Fibonacci NUmber is {}".format(fib(n-1)))
```

```python
'''
Create a simple treasure hunt game.
Create a two-dimensional 10 by 10 array of integers.
In a random position in the array store the number 1.
Repeat
  Get the user to enter coordinates where he/she think the treasure is.
  If there is a 1 at this position display 'Bingo!'.
Until they find the treasure
'''
from random import randint as rnd

array = [[0 for _ in range(10)] for k in range(10)]
x,y = rnd(0,9),rnd(0,9)

array[y][x] = 1
print(x,y)
while True:
    xu = int(input("Enter x coord (1-10):"))
    yu = int(input("Enter y coord (1-10):"))
    if array[yu-1][xu-1] == 1:
        
        print("Bingo!")
        break
    else:
        print("Try again")
        
        
```

```python
'''
Create the following Result record structure:

 HomeTeam: string
 HomeTeamScore: integer
 AwayTeam: string
 AwayTeamScore: integer

Make an array of 20 Result records.

Write a menu whose options are:

1. Add a result
2. Search for all results for a team
3. Exit

Write code to carry out these two things.

- If option 1 is selected
  get new result and
  add to the end of the results array
- If option 2 is selected
  get team name and
  display all the results that includes that team in either the home or away team

Hint: Include a variable to store the current number of results.
'''

class Result:
    def __init__(self,ht = None, hts = None, at=None, ats=None):
        self.HomeTeam = ht
        self.HomeTeamScore = hts
        self.AwayTeam =at
        self.AwayTeamScore = ats
        
records = [Result() for _ in range(20)]
pointer = 0
while True:
    option = int(input("1. Add a result \n2. Search for all results for a team \n3. Exit \n "))
    if option == 1:
        ht = input("HomeTeam: ")
        hts = input("HomeTeamScore: ")
        at = input("AwayTeam: ")
        ats = input("AwayTeamScore: ")
        records[pointer].HomeTeam, records[pointer].HomeTeamScore = ht, int(hts)
        records[pointer].AwayTeam, records[pointer].AwayTeamScore = at, int(ats)
        pointer += 1
        print("Record added")
        
    elif option == 2:
        teamName = input("Team Name here: ")
        for r in records:
            if r.HomeTeam == teamName or r.AwayTeam == teamName:
                print(r.HomeTeam, 'vs', r.AwayTeam, 'is',"{}:{}".format(r.HomeTeamScore,r.AwayTeamScore))
        
    elif option == 3:
        break
```

```python
'''
The computer generates a 4-digit code
The user types in a 4-digit code. His/her guess.
The computer tells them how many digits they guessed correctly
'''
import random
code = ''
for _ in range(4):
    code += str(random.randint(0,9))
print (code)
guess = input("enter your guess here: ")
correct = 0
for i in range(len(code)):
    if code[i] in guess:
        correct += 1
        
print("You guessed {} digit(s) correctly".format(correct))
    
```

```python
'''
For the above 4-digit code game, the computer tells them how many 
digits are guessed correctly in the correct place and how many 
digits have been guessed correctly but in the wrong place.
The user gets 12 guesses to either win â guess the right code, 
or lose â run out of guesses. 
'''
def findIndex(array,target):
    index = 0
    for e in array:
        if target == e:
            return index
        index+=1
import random
code = []
for _ in range(4):
    code += [str(random.randint(0,9))]
print (code)
win=False
for _ in range(12):
    
    guess = input("enter your guess here: ")
    guess = [x for x in guess]
    tempCode = [digit for digit in code]
    correctPlace = 0
    wrongPlace = 0
    
    for i in range(len(guess)):
        if guess[i] == tempCode[i]:
            correctPlace += 1
            tempCode[i]='x' #guessed this area correctly so i do not need to count this again 
            #print(guess[i])
            guess[i] = 'y'  #so if code is 5834 and i guess 5555 it would show as correctPlace 1 wrongPlace 0
    for j in range(len(guess)):
        if guess[j] in tempCode:
            wrongPlace += 1
            tempCode[findIndex(tempCode,guess[j])] = 'x' #similarly, code 2335 with guess of 3552 will be 3 wrong, 0 correct
            #print(guess[j])
            guess[j] = 'y'
    if correctPlace == 4:
        print("You win")
        win = True
        break
    print("Correct place: ",correctPlace)
    print("Wrong Place: ",wrongPlace)

if not win:
    print("You lose. Number was",code)
```

```python
'''
Player 1 types in a word
Player 2 has to guess the word in 5 lives
The display would look like this:

Word to guess: ******
You have 5 lives left - Letter? E
Word to guess: **EE*E
You have 5 lives left - Letter? Z
Word to guess: **EE*E
You have 4 lives left - Letter?
'''
def display(array):
    output = ''
    for a in array:
        output+=a
    return output

word = input("Player 1 word: ")
toGuess = ['*' for _ in word]
lives = 5
while True:
    if lives == 0:
        print("you lose")
        break
    if toGuess.count('*') == 0:
        print("you win")
        break

    guessed = False
    print("Word to guess:",display(toGuess))
    letter = input("You have {} lives left - Letter? ".format(lives))
    for i in range(len(toGuess)):
        if word[i] == letter:
            toGuess[i] = letter #reveal
            guessed = True
            
    if not guessed:
        lives -= 1
    
```
