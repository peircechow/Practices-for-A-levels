```python
# Write a program to work out how many days you have lived for.
import datetime as d
bday = input("Enter your birthday as dd/mm/yyyy:")
bday_split = bday.split('/')
day = int(bday_split[0])
month = int(bday_split[1])
year = int(bday_split[2])
today = d.datetime.now()

d1 = d.datetime(year, month, day,0) 
delta = today - d1 #date object - date object
print (delta.days," days has passed since your existence")
```

```python
# Write a program to work out how many seconds you have lived for.
print (int(delta.total_seconds()), " seconds has passed since your existence")
```

```python
# Write a program that will accept a date of birth and determine whether one 
# is considered an adult (i.e. 21 years old).
import datetime as d
bday = input("Enter your birthday as dd/mm/yyyy:")
bday_split = bday.split('/')
day = int(bday_split[0])
month = int(bday_split[1])
year = int(bday_split[2])
today = d.datetime.now()
adult = False
d1 = d.datetime(year, month, day,0) 
if today.year - year >= 21:
    if today.month >= month:
        adult = True

if adult:
    print("You are an adult")
else:
    print("You are not an adult.")
        

#delta = today - d1 #date object - date object
#print (delta.year," days has passed since your existence")
```
