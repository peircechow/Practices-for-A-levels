```python
# Make a game of rock, paper, scissors against the computer.
# Algorithm
# Tell user to enter either rock,paper or scissors
# Get the response
# Generate a random number from 1 to 3 (1=rock,2=paper,3=scissors)
# Compare user selection and computer selection
# Display who wins.
import random

def game(userChoice):
    status = "lose"

    cpChoice = random.randint(1,3)

    rps = {1:"rock", 2:"paper", 3: "scissors"}
    print ("Computer choice is: ", rps[cpChoice])

    if userChoice == rps[cpChoice]:

        status = "draw"

    elif userChoice == "rock":
        if rps[cpChoice] is "scissors":
            status = "win"
    elif userChoice == "scissors":
        if rps[cpChoice] is "paper":
            status ="win"
    elif userChoice == "paper":
        if rps[cpChoice] is "rock":
            status = "win"

    if status == "lose":
        print ("You lose")

    elif status == "win":
        print ("You win")

    else:
        print ("Draw")
        
    return status
    
userChoice = input("rock,paper or scissors: ")
game(userChoice)
```

```python
# For the rock, paper, scissors game above, ensure that the user enters a valid 
# entry. Add a loop structure to play several times and keep a running score.
score = 0
def isValid(a):
    if a.lower() =="rock" or a.lower()=="scissors" or a.lower() == "paper":
        return a.lower()
    return False
    
for i in range(10):
    while True:
        userChoice = input("rock,paper or scissors: ")
        if isValid(userChoice) is not False:
            break
        print("Invalid input pls try again")
    status = game(userChoice.lower())
    if status == "win":
        score += 2
    elif status == "draw":
        score += 1
    
    print ("Your score is:", score)
```

```python
# Write a program that will give the answer to logic gate questions e.g.
# Enter logic gate : OR
# Enter first input : 1
# Enter second input :0
# Result = 1
# It should work for the logic gates OR, XOR, NAND and NOR gates.
lg = input("Enter Logic gate: ")
first = input("Enter first input: ")
second = input("Enter second input: ")
result = 0

if lg == "OR":
    if first == '1' or second == '1':
        result = 1
        
elif lg == "XOR":
    result = 1
    if first == second:
        result = 0
elif lg == "NAND":
    if first == "0" or second == "0":
        result = 1
elif lg == "NOR":
    result = 1
    if first == "1" or second == "1":
        result = 0
        
print(result)
```

```python
# Write a program that will display all the factors of an input number that
# are bigger than 1.
# (e.g. the factors of the number 12 are 6,4,3 and 2 because they divide
# into 12 exactly).
# Hint: To find out whether a number X is a factor of Y use:
# If Y mod X == 0 (there is nothing remaining when Y is divided by X)
number = int(input("Enter the number here: "))

factors = []
for i in range(2,int(number**0.5+1)):
    if number % i == 0:
        if i == int(number/i): #e.g. 10,10
            factors += [i]
        else:
            factors += [i,int(number/i)]

        
print(sorted(factors))
```
