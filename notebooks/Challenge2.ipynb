```python
''' 
Write a program for a game where the computer generates a
random starting number between 20 and 30.
The player and the computer can remove 1,2 or 3 from the number
in turns. Something like this:

Starting number: 25
How many do you want to remove? 3
22 left
Computer removes 2
20 left
...

The player who has to remove the last value to bring the number
down to 0 is the loser.

1 left
Computer removes 1
You win!

Hard: Get the computer to employ a strategy to try and win. 
'''
#computer control range 4
#aim is to make the other player take his turn at 5 -> number%4==1 is the winning condition


def getNum(number):
    for i in range(1,4):
        if (number - i)%4 == 1:
            return i

number = int(input("Starting number: "))



    
while True:
    remove = int(input("How many do you want to remove? "))
    number -= remove
    if number <= 0:
        print("You Lose")
        break
   
    print (number,"left")
    if number > 4:
        comRemove = getNum(number)
    else:
        comRemove = number - 1
        
    number -= comRemove
    print("Computer removes {}".format(comRemove))
    if number <= 0:
        print("You Win!")
        break
    print (number,"left")
```

```python
'''
Write a program for a Higher / Lower guessing game.
The computer randomly generates a sequence of up to 10 numbers
between 1 and 13. The player after seeing each number in turn
has to decide whether the next number is higher or lower.
If you get 10 guesses right you win the game.

Starting number : 12
Higher(H) or lower(L)? L
Next number 8
Higher(H) or lower(L)? L
Next number 11
You lose

Hints
- Use a condition controlled loop (do until, while etc) to control the
game.
- You do not need to remember all 10 numbers, but  just the current and
the next number.
- Remember to keep a count of the number of guesses. 
'''
import random 

correct = 0
number = int(input("Starting number : "))
while correct <10:
    guess = input("Higher(H) or lower(L)? ")
    
    newNumber = random.randint(1,13)
    while newNumber == number:
        newNumber = random.randint(1,13)
        
    print ("Next number", newNumber)
    if guess == "H":
        if newNumber >number:
            correct += 1
        else:
            print ("You Lose")
            break
    elif guess == "L":
        if newNumber < number:
            correct += 1
        else:
            print ("You Lose")
            break
            
    number = newNumber
    
if correct == 10:
    print ("You win!")
```

```python
'''
Extend the above Higher / Lower number guessing game by
- Giving the player two lives
- Allowing only H or L to be entered 
'''
import random 

correct = 0
lives = 2
number = int(input("Starting number : "))
while correct <10:
    guess = input("Higher(H) or lower(L)? ")
    
    newNumber = random.randint(1,13)
    while newNumber == number:
        newNumber = random.randint(1,13)
        
    print ("Next number", newNumber)
    if guess == "H":
        if newNumber >number:
            correct += 1
        else:
            lives -= 1
            print(lives, " lives left")
    elif guess == "L":
        if newNumber < number:
            correct += 1
        else:
            lives -= 1
            print(lives, " lives left")
            
    if lives == 0:
        print("You lose")
        break
            
    number = newNumber
    
if correct == 10:
    print ("You win!")
```
